```python
n = int(input())

features, targets = [], []
for _ in range(n):
    feature, target = map(float, input().split(" "))
    features.append(feature)
    targets.append(target)

for i in range(5):
    print(f"{features[i]:.1f} {targets[i]:.1f}")

print(f"({n},2)")
```

**i) Exploratory Data Analysis (EDA)**

```python
features_mean = sum(features) / n
targets_mean = sum(targets) / n

features_std = (sum([(feature - features_mean) ** 2 for feature in features]) / n) ** 0.5
targets_std = (sum([(target - targets_mean) ** 2 for target in targets]) / n) ** 0.5

features_min = min(features)
features_max = max(features)
targets_min = min(targets)
targets_max = max(targets)

print(f"{features_mean:.2f} {features_std:.2f} {features_min:.2f} {features_max:.2f}")
print(f"{targets_mean:.2f} {targets_std:.2f} {targets_min:.2f} {targets_max:.2f}")
```

**ii) Normalization**

```python
normalized_features = []
for feature in features:
    normalized_feature = (feature - features_mean) / (features_std)
    normalized_features.append(normalized_feature)
```

**iii) Model Training**

```python
theta0 = 0.0
theta1 = 0.0
alpha = 0.01
epochs = 1000

# x -> normalized features
# y -> target variables
# y_hat -> our temporary predictions

for _ in range(epochs):

    y_hat = []
    for x in normalized_features:
        y_hat.append(theta0 + theta1 * x)       # y^ = t0 + t1 * x

    # delta0 = alpha * 1/n * sum(y_hat[i] - y[i])
    # delta1 = alpha * 1/n * sum({y_hat[i] - y[i]} * x[i])

    y_hat_minus_y = []
    for i in range(n):
        y_hat_minus_y.append(y_hat[i] - targets[i])

    y_hat_minus_y_into_x = []
    for i in range(n):
        y_hat_minus_y_into_x.append((y_hat[i] - targets[i]) * normalized_features[i])


    grad0 = alpha * 1/n * sum(y_hat_minus_y)
    grad1 = alpha * 1/n * sum(y_hat_minus_y_into_x)

    theta0 -= grad0
    theta1 -= grad1


# our final prediction after 1000 epochs
y_hat = [theta0 + theta1 * x for x in normalized_features]
# calculating difference between our pred and actual data
mse = sum((y_hat[i] - targets[i])**2 for i in range(N)) / (2*N)

print(f"Final theta0={theta0:.3f} | theta1={theta1:.3f} | Final MSE={mse:.2f}")
```

```python
import numpy as np

# Input number of data points
n = int(input().strip())

# Read data into array
arr = []
for _ in range(n):
    y = list(map(float, input().split()))
    arr.append(y)

# Show first up to 5 rows
for i in range(min(n, 5)):
    print(f"{arr[i][0]} {arr[i][1]} {arr[i][2]} {arr[i][3]}")

print(np.shape(arr))

# Separate features and target
x1, x2, x3, t = [], [], [], []
for row in arr:
    x1.append(row[0])
    x2.append(row[1])
    x3.append(row[2])
    t.append(row[3])

# Show statistics for each feature
for feature in [x1, x2, x3, t]:
    print(f"{np.mean(feature):.2f} {np.std(feature):.2f} {np.min(feature):.2f} {np.max(feature):.2f}")

# Mean and std for normalization
x1_m, x1_s = np.mean(x1), np.std(x1)
x2_m, x2_s = np.mean(x2), np.std(x2)
x3_m, x3_s = np.mean(x3), np.std(x3)

# Normalize features
x1 = [(val - x1_m) / x1_s for val in x1]
x2 = [(val - x2_m) / x2_s for val in x2]
x3 = [(val - x3_m) / x3_s for val in x3]

# Prepare matrix with bias column
mat = np.column_stack((x1, x2, x3, np.ones(n)))
mat_t = mat.T
t = np.array(t).reshape(n, 1)

# Gradient Descent
alpha = 0.01
theta = np.zeros((4, 1))
for _ in range(300):
    theta -= (alpha / n) * (mat_t @ (mat @ theta - t))

print(f"Final theta=[{theta[3,0]:.3f}, {theta[0,0]:.3f}, {theta[1,0]:.3f}, {theta[2,0]:.3f}]")

# MSE for gradient descent result
mse = np.mean(((mat @ theta - t) ** 2)) / 2
print(f"Final MSE={mse:.2f}")

# Normal equation solution
theta_x = np.linalg.inv(mat_t @ mat) @ (mat_t @ t)
mse1 = np.mean(((mat @ theta_x - t) ** 2)) / 2
print(f"MSE Difference={mse - mse1:.5f}")

# Prediction function
def predict(x1_val, x2_val, x3_val):
    x1_norm = (x1_val - x1_m) / x1_s
    x2_norm = (x2_val - x2_m) / x2_s
    x3_norm = (x3_val - x3_m) / x3_s
    features = np.array([[x1_norm, x2_norm, x3_norm, 1]])
    return (features @ theta).item()

print(f"{predict(150, 3, 5):.2f}")
print(f"{predict(200, 4, 2):.2f}")
```
